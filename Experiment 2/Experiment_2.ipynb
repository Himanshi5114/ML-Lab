```python
def candidate_elimination(X, y):
    num_attributes = len(X[0])
    S = [''] * num_attributes
    G = [['?']] * num_attributes  # Initialize G with the correct number of attributes

    for i, instance in enumerate(X):
        # Ensure instance and hypotheses have the same length
        instance = instance[:num_attributes]

        if y[i] == 'yes':
            for j in range(num_attributes):
                if S[j] == '':
                    S[j] = instance[j]
                elif S[j] != instance[j]:
                    S[j] = '?'
            # Pad hypotheses in G if necessary
            G = [g + ['?'] * (num_attributes - len(g)) for g in G]
            G = [g for g in G if all(
                g[k] == '?' or S[k] == '?' or g[k] == S[k]
                for k in range(num_attributes)
            )]
        elif y[i] == 'no':
            new_G = []
            for g in G:
                # Pad hypotheses in G if necessary
                g = g + ['?'] * (num_attributes - len(g))
                for j in range(num_attributes):
                    if g[j] == '?':
                        if S[j] != '?':
                            new_hypothesis = g.copy()
                            new_hypothesis[j] = S[j]
                            new_G.append(new_hypothesis)
            G = new_G
            G = [g for g in G if any(
                g[k] != instance[k] and g[k] != '?' for k in range(num_attributes)
            )]

    return S, G
```

```python
S_final,G_final = candidate_elimination(X,y)
print("Final Specific Hypothesis:",S_final)
print("Final General Hypothesis:",G_final)
```
